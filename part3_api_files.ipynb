## Task 2 — API Integration

```python
import requests
```

```python
# Step 1 - Fetching and displaying

BASE_URL = "https://dummyjson.com/products"

response = requests.get(f"{BASE_URL}?limit=20")
data = response.json()

products = data["products"]

print("\n--- ALL PRODUCTS ---")
print("ID  | Title                          | Category      | Price    | Rating")
print("----|--------------------------------|---------------|----------|-------")

for p in products:
    print(f"{p['id']:<3} | {p['title'][:30]:<30} | {p['category']:<13} | ${p['price']:<8} | {p['rating']}")
```

```python
# Filtering and sorting

import pandas as pd

filtered = []

# Filtering
for p in products:
    if p["rating"] >= 4.5:
        filtered.append(p)

# Sorting by price (descending)
def get_price(item):
    return item["price"]

filtered.sort(key=get_price, reverse=True)

print("\n--- FILTERED (rating >= 4.5) & SORTED BY PRICE ---")

# Converting to table
df_table = pd.DataFrame(filtered)

# Selecting only required columns
df_table = df_table[["title", "price", "rating"]]

# Renaming the columns (optional)
df_table.columns = ["Product Name", "Price ($)", "Rating"]

# Print table
print(df_table.to_string(index=False))
```

```python
# Searching by category

response = requests.get(f"{BASE_URL}/category/laptops")
laptop_data = response.json()

print("\n--- LAPTOP PRODUCTS ---")
    
# Converting list of products to DataFrame
df_laptops = pd.DataFrame(laptop_data["products"])

# Selecting only required columns
df_laptops = df_laptops[["title", "price"]]

# Renaming columns
df_laptops.columns = ["Laptop Name", "Price ($)"]

# Print table
print(df_laptops.to_string(index=False))
```

```python
# Post request and its responce

new_product = {
    "title": "My Custom Product",
    "price": 999,
    "category": "electronics",
    "description": "A product I created via API"
}

response = requests.post(f"{BASE_URL}/add", json=new_product)

print("\n--- POST RESPONSE ---")
print(response.json())
```

## Task 3 — Exception Handling

```python
# Part A — Guarded Calculator:

def safe_divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return "Error: Cannot divide by zero"
    except TypeError:
        return "Error: Invalid input types"


# Testing
print(safe_divide(10, 2))       # Normal case
print(safe_divide(10, 0))       # Zero division
print(safe_divide("ten", 2))    # Invalid type
```

```python
# Part B — Guarded File Reader:

def read_file_safe(filename):
    try:
        file = open(filename, "r", encoding="utf-8")
        content = file.read()
        file.close()
        return content
    except FileNotFoundError:
        print(f"Error: File '{filename}' not found.")
    finally:
        print("File operation attempt complete.")


# Testing
print("\nReading existing file:")
print(read_file_safe("python_notes.txt"))

print("\nReading missing file:")
print(read_file_safe("ghost_file.txt"))
```

```python
# Part C — Robust API Calls

import requests

BASE_URL = "https://dummyjson.com/products"

# Safe GET request function
def safe_get(url):
    try:
        response = requests.get(url, timeout=5)
        return response
    except requests.exceptions.ConnectionError:
        print("Connection failed. Please check your internet.")
    except requests.exceptions.Timeout:
        print("Request timed out. Try again later.")
    except Exception as e:
        print("Error:", e)
```

```python
# Safe POST request function
def safe_post(url, data):
    try:
        response = requests.post(url, json=data, timeout=5)
        return response
    except requests.exceptions.ConnectionError:
        print("Connection failed. Please check your internet.")
    except requests.exceptions.Timeout:
        print("Request timed out. Try again later.")
    except Exception as e:
        print("Error:", e)
```

```python
# Part D — Input Validation Loop:

while True:
    user_input = input("\nEnter a product ID (1–100) or 'quit': ")

    if user_input.lower() == "quit":
        print("Exiting program.")
        break

    # Validate integer input
    try:
        product_id = int(user_input)

        if product_id < 1 or product_id > 100:
            print("Please enter a number between 1 and 100.")
            continue

    except ValueError:
        print("Invalid input. Please enter a valid number.")
        continue

    # Make API call
    response = safe_get(f"https://dummyjson.com/products/{product_id}")

    if response:
        if response.status_code == 404:
            print("Product not found.")
        elif response.status_code == 200:
            data = response.json()
            print(f"Product: {data['title']} | Price: ${data['price']}")
```

## Task 4 — Logging to File 

```python
import requests
from datetime import datetime
```

```python
LOG_FILE = "error_log.txt"
```

```python
# Function to write logs

def log_error(function_name, error_type, message):
    timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

    log_entry = f"[{timestamp}] ERROR in {function_name}: {error_type} — {message}\n"

    file = open(LOG_FILE, "a", encoding="utf-8")
    file.write(log_entry)
    file.close()
```

```python
# ---------------- TEST 1: CONNECTION ERROR ----------------

def fetch_products():
    try:
        # Intentionally wrong URL to trigger ConnectionError
        requests.get("https://this-host-does-not-exist-xyz.com/api", timeout=5)
    except requests.exceptions.ConnectionError as e:
        print("Connection error occurred.")
        log_error("fetch_products", "ConnectionError", str(e))
```

```python
# ---------------- TEST 2: HTTP ERROR (404) ----------------

def lookup_product(product_id):
    try:
        url = f"https://dummyjson.com/products/{product_id}"
        response = requests.get(url, timeout=5)

        # HTTP error check (NOT exception)
        if response.status_code != 200:
            message = f"{response.status_code} Not Found for product ID {product_id}"
            print("HTTP error occurred.")
            log_error("lookup_product", "HTTPError", message)

    except Exception as e:
        # Catch any unexpected error
        log_error("lookup_product", "Exception", str(e))
```

```python
# Running test and reading log file

fetch_products()        # Will trigger ConnectionError
lookup_product(999)     # Will trigger HTTP 404


print("\n--- ERROR LOG CONTENT ---")

file = open(LOG_FILE, "r", encoding="utf-8")
print(file.read())
file.close()
```
